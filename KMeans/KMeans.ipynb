```python
import pandas as pd
data=pd.read_csv("php8vsf4S.csv")
data.columns
```

```python
happy=data['Happiness Score']
income=data['Economy (GDP per Capita)']
income.head()
```

```python
import matplotlib.pyplot as plt
```

```python
richest=data[data['Economy (GDP per Capita)']>1.4]
plt.scatter(richest['Economy (GDP per Capita)'],richest['Happiness Score'])
plt.xlabel("GDP")
plt.ylabel("happiness")
for k,row in richest.iterrows():
    
    plt.text(x=row['Economy (GDP per Capita)'],
            y=row['Happiness Score'],
            s=row['Country'])
```

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
inc=np.column_stack((income,happy))
km_res=KMeans(n_clusters=6).fit(inc)
clusters=km_res.cluster_centers_
plt.scatter(income,happy)
plt.scatter(clusters[:,0],clusters[:,1],s=100)
```
