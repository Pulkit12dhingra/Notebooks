```python
import pandas as pd
data=pd.read_csv("Kmeans.csv")
data.columns
```

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
data.head()
```

```python
inc=np.column_stack((data["V1"],data["V2"]))
km_res=KMeans(n_clusters=8).fit(inc)
clusters=km_res.cluster_centers_
plt.scatter(data["V1"],data["V2"])
plt.scatter(clusters[:,0],clusters[:,1],s=100)
```
